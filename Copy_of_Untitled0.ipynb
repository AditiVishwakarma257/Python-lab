[1,2,3,4,5] → These are your x-values.

[1,4,9,16,25] → These are your y-values (squares of x-values).

'r^-' → This is a format string that styles the plot:

'r' → Red color

'^' → Triangle-up markers for each point

'-' → Solid line connecting the points

```python
plt.grid(True)
plt.title('squares of the numbers')
plt.xlabel('The Numbers', fontsize=18, color='red')
plt.ylabel('The squares',fontsize=18, color='red')
plt.plot([1,2,3,4,5], [1,4,9,16,25], 'r^-', linewidth=3.0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Set a large figure size for the 'large canvas' effect and specific light gray background
fig = plt.figure(figsize=(12, 10), facecolor='#f2f2f2')
ax = fig.add_subplot(111, projection='3d')

# Generate random data for scatter points (between 350 and 450 points)
num_points = np.random.randint(350, 451) # Random number of points
x_scatter = np.random.rand(num_points)
y_scatter = np.random.rand(num_points)
z_scatter = np.random.rand(num_points)

# Plot the scatter points: small, soft light-blue, slightly transparent
ax.scatter(x_scatter, y_scatter, z_scatter, color='skyblue', s=20, alpha=0.6, edgecolors='none')

# Set axis limits to create a normalized space (0 to 1) for the coordinate system
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)

# Add thin black X, Y, Z axes meeting at one corner
# These explicitly drawn lines define the visual axes from the origin.
ax.plot([0, 1], [0, 0], [0, 0], color='black', linewidth=0.8) # X-axis
ax.plot([0, 0], [0, 1], [0, 0], color='black', linewidth=0.8) # Y-axis
ax.plot([0, 0], [0, 0], [0, 1], color='black', linewidth=0.8) # Z-axis

# Set axis labels
ax.set_xlabel('X-axis', color='black')
ax.set_ylabel('Y-axis', color='black')
ax.set_zlabel('Z-axis', color='black')

# Remove grid lines
ax.grid(False)

# Make panes completely transparent (no background color for the planes)
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False

# Set pane edge colors to white or transparent to effectively remove them visually
ax.xaxis.pane.set_edgecolor('w')
ax.yaxis.pane.set_edgecolor('w')
ax.zaxis.pane.set_edgecolor('w')

# Remove axis ticks and tick labels for a minimalist look
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# Adjust the view angle as specified
ax.view_init(elev=18, azim=235)

# No title, legends, branding, or extra text
plt.title('')
plt.figtext(0.5, 0.01, '', wrap=True, horizontalalignment='center', fontsize=12) # Clear any default figtext

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- General Figure Settings ---
# Light gray overall background
fig = plt.figure(figsize=(10, 6), facecolor='#f5f5f5')

# Generate data for smooth curves (500+ data points)
x = np.linspace(0, 2 * np.pi, 500)

# --- Subplot Configuration Function ---
def configure_subplot(ax, title, xlabel='X-axis', ylabel='Y-axis'):
    ax.set_title(title, fontsize=12, color='dimgray')
    ax.set_xlabel(xlabel, fontsize=10, color='gray')
    ax.set_ylabel(ylabel, fontsize=10, color='gray')

    # Subplot background color slightly darker than figure background
    ax.set_facecolor('#ececec')

    # Add light grid lines
    ax.grid(True, linestyle='--', alpha=0.6, color='#d3d3d3', linewidth=0.7)

    # Remove top and right spines for modern UI appearance
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('lightgray')
    ax.spines['bottom'].set_color('lightgray')
    ax.tick_params(colors='dimgray') # Make ticks slightly darker

# --- Top-Left Subplot: sin(x) ---
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, np.sin(x), color='orange', linestyle='-', linewidth=2)
configure_subplot(ax1, 'sin(x)')

# --- Top-Right Subplot: cos(x) ---
ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x, np.cos(x), color='magenta', linestyle='-', linewidth=2)
configure_subplot(ax2, 'cos(x)')

# --- Bottom-Left Subplot: sin²(x) ---
ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(x, np.sin(x)**2, color='orange', linestyle='--', linewidth=2)
configure_subplot(ax3, 'sin²(x)')

# --- Bottom-Right Subplot: cos²(x) ---
ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(x, np.cos(x)**2, color='magenta', linestyle='--', linewidth=2)
configure_subplot(ax4, 'cos²(x)')

# Use plt.tight_layout()
plt.tight_layout(pad=2.0) # Add some padding between subplots

plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_true = [0, 1, 0, 1, 0, 1]
y_pred = [0, 1, 0, 0, 0, 1]

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

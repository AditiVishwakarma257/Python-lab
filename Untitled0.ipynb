```python
n = int(input("Enter number: "))

fact = 1
sum_fact = 0

for i in range(1, n + 1):
    fact = fact * i      # calculate i!
    sum_fact += fact     # add to sum

print("Sum of factorial series is:", sum_fact)
```

```python
max_val = 0

while True:
    i = input("Enter number (type NO to stop): ")

    if i == "NO":
        break

    i = int(i)   # convert to integer

    if i > max_val:
        max_val = i

print("Maximum is", max_val)
```

```python
class Car:
    def __init__(self, make, model, year, mileage=0):
        self.make = make
        self.model = model
        self.year = year
        self.mileage = mileage

    def drive(self, distance):
        self.mileage += distance
        print(f"{self.make} {self.model} drove {distance} km")

    def show_details(self):
        print(f"Car: {self.year} {self.make} {self.model}")
        print(f"Mileage: {self.mileage} km")


# Creating object
car1 = Car("Toyota", "Fortuner", 2022, 15000)

# Using methods
car1.show_details()
car1.drive(100)
car1.show_details()
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def get_marks(self):
        return self.marks


# Creating object
s1 = Student("Aditi", 95)

# Getting marks
print("Marks:", s1.get_marks())
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def annual_salary(self):
        return self.salary * 12


# list of objects
employees = [
    Employee("Aditi", 25000),
    Employee("Rahul", 30000),
    Employee("Neha", 28000)
]

# display
for emp in employees:
    print(emp.name, emp.salary, emp.annual_salary())
```
